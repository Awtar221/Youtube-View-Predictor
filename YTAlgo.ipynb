```python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Download latest version
path = kagglehub.dataset_download("abdulwadood11220/youtube-data-for-analytics-600-rows")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv('/kaggle/input/youtube-data-for-analytics-600-rows/youtube_data.csv')
```

```python
df.head()
```

```python
df['ispop'] = df['view_count'] > 1000000
df.head()
```

```python
df.describe()
```

```python
import re

def parse_duration(duration_str):
    # Example: "PT1H2M3S", "PT4M", "PT30S"
    hours = minutes = seconds = 0
    match = re.match(r'PT(?:(\d+)H)?(?:(\d+)M)?(?:(\d+)S)?', duration_str)
    if match:
        hours = int(match.group(1)) if match.group(1) else 0
        minutes = int(match.group(2)) if match.group(2) else 0
        seconds = int(match.group(3)) if match.group(3) else 0
    return hours * 3600 + minutes * 60 + seconds
```

```python
df['duration'] = df['duration'].apply(parse_duration)
```

```python
df.head()
```

```python
try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
  pass
import tensorflow as tf
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import TextVectorization
from sklearn.model_selection import train_test_split


X_numeric = df[['duration', 'like_count']]
X_text_title = df['title']
X_text_tags = df['tags']
y = df['view_count']

X_train_num, X_test_num, X_train_title, X_test_title, X_train_tags, X_test_tags, y_train, y_test = train_test_split(
    X_numeric, X_text_title, X_text_tags, y, test_size=0.2, random_state=42
)

max_tokens = 10000
output_seq_len = 20

title_vectorizer = TextVectorization(max_tokens=max_tokens, output_sequence_length=output_seq_len)
title_vectorizer.adapt(X_train_title)
tags_vectorizer = TextVectorization(max_tokens=max_tokens, output_sequence_length=output_seq_len)
tags_vectorizer.adapt(X_train_tags)
```

```python
numeric_input = tf.keras.Input(shape=(X_numeric.shape[1],), name="numeric_input")
title_input = tf.keras.Input(shape=(1,), dtype=tf.string, name="title_input")
tags_input = tf.keras.Input(shape=(1,), dtype=tf.string, name="tags_input")
```

```python
x_title = title_vectorizer(title_input)
x_title = layers.Embedding(input_dim=max_tokens, output_dim=16)(x_title)
x_title = layers.GlobalAveragePooling1D()(x_title)

x_tags = tags_vectorizer(tags_input)
x_tags = layers.Embedding(input_dim=max_tokens, output_dim=16)(x_tags)
x_tags = layers.GlobalAveragePooling1D()(x_tags)
```

```python
x_num = layers.Dense(32, activation='relu')(numeric_input)
```

```python
# Inspect shapes and dtypes of text inputs after vectorization and embedding
print("Shape and dtype of x_title:", x_title.shape, x_title.dtype)
print("Shape and dtype of x_tags:", x_tags.shape, x_tags.dtype)

x = layers.concatenate([x_title, x_tags, x_num])
x = layers.Dense(64, activation='relu')(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(1)(x)
```

```python
model = keras.Model(inputs=[title_input, tags_input, numeric_input], outputs=x)
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
%%time
print("y_train dtype:", y_train.dtype)
print("y_test dtype:", y_test.dtype)
history = model.fit(
    {
        "title_input": X_train_title.to_numpy(),
        "tags_input": X_train_tags.to_numpy(),
        "numeric_input": X_train_num
    },
    y_train,
    epochs=50,
    validation_data=(
        {
            "title_input": X_test_title.to_numpy(),
            "tags_input": X_test_tags.to_numpy(),
            "numeric_input": X_test_num
        },
        y_test
    )
)
```

```python
loss, mae = model.evaluate(
    {"title_input": X_test_title.to_numpy(),"tags_input" : X_test_tags.to_numpy(), "numeric_input": X_test_num},
    y_test
)
print(f"Loss: {loss}, MAE: {mae}")
```

```python
log_pred = model.predict({
    "title_input": tf.constant(["10$ Phone VS 100$ Phone"]),
    "tags_input": tf.constant(['["tech", "gadgets", "innovation"]']),
    "numeric_input": tf.constant([[15, 22]])
})[0][0]

predicted_views = log_pred
print(log_pred)
```
